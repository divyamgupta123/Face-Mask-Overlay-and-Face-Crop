```python
import cv2
import dlib
import numpy as np
import os
import imutils
import matplotlib.pyplot as plt
import csv
from PIL import Image
```

## Crop face region

```python
def detect_faces(image):
    # Create a face detector
    face_detector = dlib.get_frontal_face_detector()

    # Run detector and get bounding boxes of the faces on image.
    detected_faces = face_detector(image, 1)
    face_frames = [(x.left(), x.top(),
                    x.right(), x.bottom()) for x in detected_faces]


    return face_frames
```

```python
def cropFace(path):
    
    img_path = path

    #read image
    image = cv2.imread(img_path)
    
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    height = image.shape[0]
    width = image.shape[1]

    # Detect faces
    detected_faces = detect_faces(image)

    # Crop faces and plot
    for n, face_rect in enumerate(detected_faces):
        startX = face_rect[0]
        startY = face_rect[1]
        endX = face_rect[2]
        endY = face_rect[3]
        
        face = Image.fromarray(image).crop((startX,startY,endX,endY))
        
        fig = plt.figure()
        plt.subplot(121)
        plt.imshow(image)

        plt.subplot(122)
        plt.imshow(face)
        plt.show()
        
        fig.savefig('./results/faceCrop.jpg')
```

```python
cropFace('./test images/000582.jpg')
```

## Applying mask on face

```python
# Path of file
p = "./dlib files/shape_predictor_68_face_landmarks.dat"
# Initialize dlib's shape predictor
predictor = dlib.shape_predictor(p)
```

```python
def getMaskImageArr(mask_image_path):
    mask_img = cv2.imread(mask_image_path, cv2.IMREAD_UNCHANGED)
    mask_img = mask_img.astype(np.float32)
    mask_img = mask_img/255.0
    return mask_img
```

```python
def getMaskSrcPts(mask_csv_path):
    with open(mask_csv_path) as csv_file:
      csv_reader = csv.reader(csv_file, delimiter=",")
      src_pts = []
      for i, row in enumerate(csv_reader):
          # skip head or empty line if it's there
          try:
              src_pts.append(np.array([float(row[1]), float(row[2])]))
          except ValueError:
              continue
    src_pts = np.array(src_pts, dtype="int")
    return src_pts
```

```python
def getImageDstPts(path):
    # Loading the image and resizing, converting it to grayscale
    img= cv2.imread(path)

    img = imutils.resize(img, width = 500)

    # Initialize dlib's face detector
    detector = dlib.get_frontal_face_detector()

    """
    Detecting faces in the grayscale image and creating an object - faces to store the list of bounding rectangles coordinates
    The "1" in the second argument indicates that we should upsample the image 1 time.  
    This will make everything bigger and allow us to detect more faces
    """

    faces = detector(img, 1)

    # printing the coordinates of the bounding rectangles
    print("Number of faces detected: ", len(faces))

    dst_pts = []
    for face in faces:
        landmarks = predictor(img, face)


    #Covering the face area starting from landmark 2 till 16 and landmark 30
    #landmarks are according to the facial landmark image in the images folder
        
    for i in range(1, 16):
        point = [landmarks.part(i).x, landmarks.part(i).y]
        dst_pts.append(point)

    dst_pts.append([landmarks.part(29).x, landmarks.part(29).y])

    dst_pts = np.array(dst_pts)

    return img,dst_pts
```

```python
def faceMaskImage(img,src_pts,dst_pts,mask_image_arr):
    result = img.copy()
    
    result = result.astype(np.float32) / 255.0

    # get the perspective transformation matrix
    M, _ = cv2.findHomography(src_pts, dst_pts)

    # transformed masked image
    transformed_mask = cv2.warpPerspective(
        mask_image_arr,
        M,
        (result.shape[1], result.shape[0]),
        None,
        cv2.INTER_LINEAR,
        cv2.BORDER_CONSTANT,
    )

    # mask overlay
    alpha_mask = transformed_mask[:, :, 3]
    alpha_image = 1.0 - alpha_mask


    for c in range(0, 3):
        result[:, :, c] = (
            alpha_mask * transformed_mask[:, :, c]
            + alpha_image * result[:, :, c]
        )

    result=cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
    
    return result
```

```python
src_pts = getMaskSrcPts('./masks/3m.csv')
img,dst_pts = getImageDstPts('./test images/000582.jpg')
mask_images_arr = getMaskImageArr('./masks/3m.png')
result = faceMaskImage(img,src_pts,dst_pts,mask_images_arr)

img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

fig = plt.figure()
plt.subplot(121)
plt.imshow(img)

plt.subplot(122)
plt.imshow(result)
plt.show()

fig.savefig('./results/mask3m.jpg')
```
